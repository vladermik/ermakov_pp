```python

import random
fruits = ['', '', '', '', '', '', '', '', '', '']
counts = [509, 335, 39, 38, 37, 24, 23, 22, 22, 19]

all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=11)

# Plot Bars
plt.figure(figsize=(20,7), dpi= 80)
plt.bar(fruits, counts, color=c, width=.5)
for i, val in enumerate(counts):
    plt.text(i, val, int(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

# Decoration
plt.gca().set_xticklabels(fruits, rotation=60, horizontalalignment= 'right')
plt.title("  ,   ", fontsize=22)
plt.ylabel('')
plt.ylim(0, 509)
plt.show()
```
